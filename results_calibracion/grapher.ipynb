```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar CSV
df = pd.read_csv("calibracion_countsketch_31mer.csv")
df.sort_values(by=['tamano'], inplace=True)
# df = df[df['tamano'] > 60000].copy()


plt.figure(figsize=(8,6))

# Línea de MAE
plt.plot(
    df['tamano'],
    df['mae'],
    marker='o',
    linestyle='-',
    label="MAE"
)

# Línea de MRE
plt.plot(
    df['tamano'],
    df['mre'],
    marker='s',
    linestyle='-',
    label="MRE"
)

plt.title("MAE y MRE vs tamaño del CountSketch")
plt.xlabel("Tamaño en bytes")
plt.ylabel("Error")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()
plt.savefig("calibracion_countsketch_31mer.png")
plt.show()
```

```python
# Cargar CSV
df = pd.read_csv("calibracion_towersketch_31mer.csv")
df.sort_values(by=['tamano'], inplace=True)
df['tamano'].drop_duplicates(inplace=True)
df = df[df['mae'] < 1000].copy()
df = df[df['d'] == 7].copy()


plt.figure(figsize=(8,6))

# Línea de MAE
plt.plot(
    df['tamano'],
    df['mae'],
    marker='o',
    linestyle='-',
    label="MAE"
)

# Línea de MRE
plt.plot(
    df['tamano'],
    df['mre'],
    marker='s',
    linestyle='-',
    label="MRE"
)

plt.title("MAE y MRE vs tamaño del TowerSketch")
plt.xlabel("Tamaño en bytes")
plt.ylabel("Error")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()
plt.savefig("calibracion_towersketch_31mer.png")
plt.show()
```
